### Область работы 2 (поиск  модели .... )

```python
# Данный блок(и) НЕ выполняются преподавателем в области 2
# блок(и) предназначены для поиска лучшей модели 
# должен быть понятен и очевиден отбор параметров модели
# 
# Запрещается размещать данные блоки за пределами обасти 2
# Все блоки данной области должны быть выполнены
#
# ЗАПРЕЩАЕТСЯ ИСПОЛЬЗОВАТЬ ТЕСТОВЫЙ НАБОР
#
# Путь к тренировочному набору
# 

path_train = 'train.csv' # содержит только имя файла, без имен папок !!!
```

```python
df = pd.read_csv(path_train)
display(df.head())
display(df.describe())
display(df.info())
```

```python
df.drop_duplicates(inplace=True)
```

```python
missing_values = df.isnull().sum()
print("Количество пропущенных значений по признакам:")
print(missing_values)
```

```python
X = df.drop(['Weight'], axis = 1)
y = df['Weight']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 52)
X_train.info()
```

```python
numeric_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = df.select_dtypes(include=['object', 'category']).columns.tolist()

print(f"Numeric features: {numeric_features}")
print(f"Categorical features: {categorical_features}")
```

```python
cat_features = ['Age','Sex']
num_features = ['BCS', 'Length', 'Girth', 'Height']
```

```python
age = ['Age']
num_f = ['Length', 'Girth', 'Height']
num = ['BCS']
sex = ['Sex']
```

```python
age_map = [{
    'col':'Age',
    'mapping':{'>20':5, '15-20':4, '10-15':3, '5-10':2, '2-5':1, '<2':0}
    }]
```

```python
transformer_age = Pipeline(steps=[
    ('ce',ce.OrdinalEncoder(mapping=age_map)),
    ('scaler', StandardScaler())
])

transformer_sex = Pipeline(steps=[
    ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore').set_output(transform='pandas')),
    ('scaler', StandardScaler())
])

transformer_num_f = Pipeline(steps=[
    ('polynom', PolynomialFeatures(3,include_bias=False)),
    ('scaler', StandardScaler())
])

transformer_num = Pipeline(steps=[
    ('scaler', StandardScaler())
])
```

```python
CT = ColumnTransformer([
        ("age", transformer_age, age),
        ("num", transformer_num, num),
        ("sex", transformer_sex, sex),
        ("num_f", transformer_num_f, num_f)
        ]).set_output(transform="pandas")

display(CT)
```

```python
gb = Pipeline([('preprocessing', CT), ('regressor', GradientBoostingRegressor(learning_rate=0.01,
                                                                              subsample=0.1,
                                                                              min_samples_split = 9,
                                                                              max_depth = 4,
                                                                              n_estimators=500,
                                                                              random_state=52))])
gb.fit(X_train, y_train)
y_pred_test = gb.predict(X_test)
y_pred_train = gb.predict(X_train)

print('Правильность на тренировочном наборе:', mean_squared_error(y_train, y_pred_train))
print('Правильность на тестовом наборе:', mean_squared_error(y_test, y_pred_test))
```

```python
rf = Pipeline([('preprocessing', CT), ('regressor', RandomForestRegressor(n_estimators=500, 
                                                                           max_depth=4, 
                                                                           min_samples_split = 9, 
                                                                           min_samples_leaf = 5, 
                                                                           max_features=None, 
                                                                           random_state=42, 
                                                                           n_jobs=-1))])
rf.fit(X_train, y_train)
y_pred_test = rf.predict(X_test)
y_pred_train = rf.predict(X_train)

print('Правильность на тренировочном наборе:', mean_squared_error(y_train, y_pred_train))
print('Правильность на тестовом наборе:', mean_squared_error(y_test, y_pred_test))
```

```python
pipe = Pipeline([('preprocessing', CT), 
                 ('classifier', LinearRegression())])

regressors = [Ridge(), ElasticNet(), Lars(), Lasso(), LinearRegression()]

param_grid = [
    {'classifier': [Ridge()],
     'classifier__alpha': [0.01, 0.1]},
    {'classifier': [ElasticNet()],
     'classifier__alpha': [0.01, 0.1],
     'classifier__l1_ratio': [0.1, 0.5, 0.9]},
    {'classifier': [Lars()],
     'classifier__n_nonzero_coefs': [1, 5, 10]},
    {'classifier': [Lasso()],
     'classifier__alpha': [0.01, 0.1]},
    {'classifier': [LinearRegression()]},
]

grid = GridSearchCV(pipe, param_grid, cv = KFold(5), return_train_score = True, scoring = 'neg_mean_squared_error', n_jobs=-1)
grid.fit(X_train, y_train)
print("Наилучшие значения параметров: {}".format(grid.best_params_))
print("Наилучшее значение кросс-валидац. правильности:{:.2f}".format(grid.best_score_))
print("Правильность для наилучшей модели на тестовом наборе: {:.6f}\n".format(grid.score(X_test, y_test)))
```
