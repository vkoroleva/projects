### Область работы 2 (выполнение лучшей модели)

```python
# Данный блок(и) в области 2 выполняется преподавателем
#
# В области находится одна, единственная, итоговая модель машинного обучения с однозначными, 
# зафиксированными параметрами
#
# В данной области категорически запрещается искать, выбирать, улучшать, оптимизировать, 
# тюниговать и т.д. модель машинного обучения
```

```python
# Путь к тренировочному набору
path_train = 'train.csv' # содержит только имя файла, без имен папок
path_train_y = 'train_Y.csv'
# Путь к тестовому набору
path_test  = 'test.csv' # содержит только имя файла, без имен папок
```

```python
X_train = pd.read_csv(path_train)
y_train = pd.read_csv(path_train_y)
X_test = pd.read_csv(path_test)
```

```python
X_train.drop_duplicates(inplace=True)
y_train = y_train.loc[X_train.index]
```

```python
y_train = y_train['Diabetes_binary']
```

```python
# Блок обучения модели (все три модели)
```

```python
rf = RandomForestClassifier(n_estimators=200, max_depth=10, min_samples_split = 10, min_samples_leaf = 10, max_features=None, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
ftc_rf = FixedThresholdClassifier(rf, threshold=0.32030701410104256, response_method="predict_proba")
ftc_rf.fit(X_train, y_train)
```

```python
gb = GradientBoostingClassifier(learning_rate=0.1, subsample=0.75, n_estimators=200,random_state=42)
gb.fit(X_train, y_train)
ftc_gb = FixedThresholdClassifier(gb, threshold=0.31556538937073886, response_method="predict_proba")
ftc_gb.fit(X_train, y_train)
```

```python
gb = GradientBoostingClassifier(learning_rate=0.1, subsample=0.75, n_estimators=200,random_state=42)
gb.fit(X_train, y_train)
ftc_gb = FixedThresholdClassifier(gb, threshold=0.31556538937073886, response_method="predict_proba")
ftc_gb.fit(X_train, y_train)
```

```python
# Блок предсказания с использованием тестового набора (все три модели)
```

```python
y_pred_RF = ftc_rf.predict(X_test)
y_pred_GB = ftc_gb.predict(X_test)
y_pred_final = ftc_gb.predict(X_test)
```

```python
# Вектора предсказанных значений  RandomForestClassifier или ExtraTreesClassifier
y_predict_RF = y_pred_RF
```

```python
# Вектора предсказанных значений  GradientBoostingClassifier или HistGradientBoostingClassifier
y_predict_GB = y_pred_GB
```

```python
# Вектора предсказанных значений  финальной модели
y_predict_final = y_pred_final
```
