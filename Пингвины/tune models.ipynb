### Область работы 2 (поиск  модели .... )

```python
# Данный блок(и) НЕ выполняются преподавателем в области 2
# блок(и) предназначены для поиска лучшей модели 
# должен быть понятен и очевиден отбор параметров модели
# 
# Запрещается размещать данные блоки за пределами обасти 2
# Все блоки данной области должны быть выполнены
#
# ЗАПРЕЩАЕТСЯ ИСПОЛЬЗОВАТЬ ТЕСТОВЫЙ НАБОР
#
# Путь к тренировочному набору
# 

path_train = 'train.csv' # содержит только имя файла, без имен папок !!!
```

```python
df = pd.read_csv(path_train)
display(df.head())
display(df.describe())
display(df.info())
```

```python
df.drop_duplicates(inplace=True)
```

```python
missing_values = df.isnull().sum()
print("Количество пропущенных значений по признакам:")
print(missing_values)
```

```python
X = df.drop(['species'], axis = 1)
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 52,
                                                    stratify = y)
X_train.info()
```

```python
transformer_cat = Pipeline(steps=[
    ('imputer', SimpleImputer(missing_values=np.nan, strategy='most_frequent', add_indicator=False)),
    ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore').set_output(transform='pandas')),
    ('scaler', StandardScaler())
])

transformer_num = Pipeline(steps=[
    ('imputer', SimpleImputer(missing_values=np.nan, strategy='mean', add_indicator=False)),
    ('scaler', StandardScaler())
])
```

```python
cat_features = ['island','sex']
num_features = ['bill_depth_mm', 'bill_length_mm', 'body_mass_g', 'flipper_length_mm']
```

```python
CT = ColumnTransformer([
        ("num", transformer_num, num_features),
        ("cat", transformer_cat, cat_features),
        ]).set_output(transform="pandas")

display(CT)
```

```python
pipe_LR = Pipeline([('preprocessing', CT),
                 ('clf', LogisticRegression(max_iter = 7000))])

cv_LR = StratifiedShuffleSplit(test_size = 0.2, n_splits = 5, random_state = 42)

param_grid_LR =[
    {'clf__penalty': ['l2'],
     'clf__solver': ['newton-cg' ,'lbfgs', 'liblinear', 'sag', 'saga'],
    },
    {'clf__penalty': ['l1'],
     'clf__solver': ['liblinear','saga'],
    },
    {'clf__penalty': [None],
     'clf__solver': ['newton-cg' ,'lbfgs','sag', 'saga'],
    }
    ]

grid_LR = GridSearchCV(pipe_LR, param_grid_LR, scoring = 'accuracy', cv = cv_LR, n_jobs=-1)
grid_LR.fit(X_train, y_train)
print("Наилучшие значения параметров: {}".format(grid_LR.best_params_))
print("Наилучшее значение кросс-валидац. правильности:{:.2f}".format(grid_LR.best_score_))
```

```python
print("Правильность для наилучшей модели на тестовом наборе: {:.6f}\n".format(grid_LR.score(X_test, y_test)))
```

```python
pipe_svm = Pipeline([('preprocessing', CT),
                 ('classifier', svm.SVC(kernel = 'linear', probability = True))])

cv_svm = StratifiedShuffleSplit(test_size = 0.2, n_splits = 5, random_state = 42)

C = [0.0001, 0.001, 0.01, 0.1]
gamma = ['scale', 'auto', 0.01, 0.1, 1]

param_grid_svm =[
    {'classifier': [svm.SVC()],
     'classifier__C': C,
     'classifier__gamma': gamma
     }
]

grid_svm = GridSearchCV(pipe_svm, param_grid_svm, scoring = 'accuracy', cv = cv_svm, n_jobs=-1)
grid_svm.fit(X_train,y_train)
print("Наилучшие значения параметров: {}".format(grid_svm.best_params_))
print("Наилучшее значение кросс-валидац. правильности:{:.2f}".format(grid_svm.best_score_))
```

```python
print("Правильность для наилучшей модели на тестовом наборе: {:.6f}\n".format(grid_svm.score(X_test, y_test)))
```
